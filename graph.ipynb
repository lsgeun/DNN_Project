# step, tanh, sigmoid, relu 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0.01*i for i in range(-500, 500+1)])
# step 그래프
step = np.array(x > 0, dtype='int32')

plt.plot(x, step, linewidth=4)
plt.plot([0,0], [step.min(), step.max()], "k")
plt.plot(0, step.max(), "k^")
plt.plot([x.min(), x.max()], [0,0], "k")
plt.plot(x.max(), 0, "k>")

plt.xlabel("x")
plt.ylabel("y=step(x)")
plt.savefig("/Users/isgeun/Library/Mobile Documents/com~apple~CloudDocs/INBOX/step.png")
plt.show()
# tanh 그래프
tanh = np.tanh(x)

plt.plot(x, tanh, linewidth=4)
plt.plot([0,0], [tanh.min(), tanh.max()], "k")
plt.plot(0, tanh.max(), "k^")
plt.plot([x.min(), x.max()], [0,0], "k")
plt.plot(x.max(), 0, "k>")

plt.xlabel("x")
plt.ylabel("y=tanh(x)")
plt.savefig("/Users/isgeun/Library/Mobile Documents/com~apple~CloudDocs/INBOX/tanh.png")
plt.show()
# Sigmoid 그래프
sigmoid = 1/(1+np.exp(-x))

plt.plot(x, sigmoid, linewidth=4)
plt.plot([0,0], [sigmoid.min(), sigmoid.max()], "k")
plt.plot(0, sigmoid.max(), "k^")
plt.plot([x.min(), x.max()], [0,0], "k")
plt.plot(x.max(), 0, "k>")

plt.xlabel("x")
plt.ylabel("y=sigmoid(x)")
plt.savefig("/Users/isgeun/Library/Mobile Documents/com~apple~CloudDocs/INBOX/sigmoid.png")
plt.show()
# Relu 그래프

Relu = x.copy()
nonpositive_mask = Relu <= 0
np.place(Relu, nonpositive_mask, 0)

plt.plot(x, Relu, linewidth=4)
plt.plot([0,0], [Relu.min(), Relu.max()], "k")
plt.plot(0, Relu.max(), "k^")
plt.plot([x.min(), x.max()], [0,0], "k")
plt.plot(x.max(), 0, "k>")

plt.xlabel("x")
plt.ylabel("y=Relu(x)")
plt.savefig("/Users/isgeun/Library/Mobile Documents/com~apple~CloudDocs/INBOX/relu.png")
plt.show()
```
