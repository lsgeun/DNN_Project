# 대용량 넘파이 크기

```python
import numpy as np

KB = 2**10
MB = 2**20
GB = 2**30

# 예상되는 데이터의 크기의 maximum이 (max_data_count, max_row, max_column)인데
# 이때 데이터 타입 별로 물리적 데이터 크기를 구함.
max_data_count = 90000
max_row = 100
max_column = 100
```

```python
int_32 = np.zeros((max_data_count, max_row, max_column), dtype="int32")
print("int_32:", int_32.nbytes/KB, "KB,", int_32.nbytes/MB, "MB,", int_32.nbytes/GB, "GB,")
del int_32 # RAM을 많이 먹기 때문에 크기만 출력하고 지움
```

```python
int_64 = np.zeros((max_data_count, max_row, max_column), dtype="int64")
print("int_64:", int_64.nbytes/KB, "KB,", int_64.nbytes/MB, "MB,", int_64.nbytes/GB, "GB,")
del int_64 # 마찬가지로 RAM을 많이 먹기 때문에 크기만 출력하고 지움
```

```python
float_16 = np.zeros((max_data_count, max_row, max_column), dtype="float16")
print("float_16:", float_16.nbytes/KB, "KB,", float_16.nbytes/MB, "MB,", float_16.nbytes/GB, "GB,")
del float_16 # 마찬가지로 RAM을 많이 먹기 때문에 크기만 출력하고 지움
```

```python
float_32 = np.zeros((max_data_count, max_row, max_column), dtype="float32")
print("float_32:", float_32.nbytes/KB, "KB,", float_32.nbytes/MB, "MB,", float_32.nbytes/GB, "GB,")
del float_32 # 마찬가지로 RAM을 많이 먹기 때문에 크기만 출력하고 지움
```

```python
float_64 = np.zeros((max_data_count, max_row, max_column), dtype="float64")
print("float_64:", float_64.nbytes/KB, "KB,", float_64.nbytes/MB, "MB,", float_64.nbytes/GB, "GB,")
del float_64 # 마찬가지로 RAM을 많이 먹기 때문에 크기만 출력하고 지움
```

# 각 레이어에서 행과 열 추출

```python
import numpy as np

layer_0 = np.array([1,2]).reshape(1,2) # = reshape(1,-1)
layer_1 = np.array([10,23]).reshape(1,2)
layer_2 = np.array([24,3]).reshape(1,2)
layer_3 = np.array([132,83]).reshape(1,2)
layer_4 = np.array([9,32]).reshape(1,2)

layers = np.append(layer_0, layer_1, axis=0)
layers = np.append(layers, layer_2, axis=0)
layers = np.append(layers, layer_3, axis=0)
layers = np.append(layers, layer_4, axis=0)

layers = layers.T

N = layers[0]
M = layers[1]

NM = np.append(N, M)
print(NM)
```
