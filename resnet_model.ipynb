3. stl-10 데이터 셋 가지고 오기

```python
# loading training data
import torchvision.transforms as transforms
from torchvision.datasets import STL10
import os

path2data = './data'

# if not exists the path, make the path
if not os.path.exists(path2data):
    os.mkdir(path2data)

data_transformer = transforms.Compose([transforms.ToTensor()])
train_ds = STL10(path2data, split='train', download='True', transform=data_transformer)

print(train_ds.data.shape)
```

```python
import collections
y_train = [y for _,y in train_ds]
counter_train = collections.Counter(y_train)
print(counter_train)
```

```python
# load test dataset and call it test0_ds
test0_ds = STL10(path2data, split='test', download=True, transform=data_transformer)
print(test0_ds.data.shape)
```
